```python
import time

import numpy as np
from geneticalgorithm import geneticalgorithm as ga
from matplotlib import pyplot as plt


def rosenbrock(X):
    return sum([100 * (X[i + 1] - X[i] ** 2) ** 2 + (1 - X[i]) ** 2
                for i in range(len(X) - 1)])



varbound = np.array([[-10,10]]*3)  
model = ga(function=rosenbrock, dimension=3, variable_type='int', variable_boundaries=varbound)
start = time.time()
model.run()
elapsed_time = time.time() - start




print("=========================================")
print("Массив переменных для генетического алгоритма.: ",varbound)
print("Количество итераций:", model.iterate)
print("Минимальное значение дисперсии:", model.best_function)
print("Оптимальные значения параметров:", model.best_variable)

print("Затраченное время: {} сек".format(elapsed_time))
print("=========================================")



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-2, 2, 100)
y = np.linspace(-1, 3, 100)
X, Y = np.meshgrid(x, y)
Z = rosenbrock([X, Y])

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('Функция Розенброка')
plt.show()

```
